```python
from typing import TypedDict, List, Annotated
from langgraph.graph import StateGraph, END
```

```python
class State(TypedDict):
    count:int
    # histories:List[str]
```

```python
def increment(state: State) -> State:
    return {
        "count":state["count"]+1,
        # "histories":state["histories"]+["incremented"]
    }
```

```python
def route(state):
    if state["count"]<3:
        return "increment"
    return "stop"
```

```python
initial_state: State = {
    "count":0,
    # "histories":[]    
}
```

```python
graph = StateGraph(State)
```

```python
graph.add_node("increment", increment)
# graph.add_node("route", route)
# graph.add_node("decrement", decrement)
graph.set_entry_point("increment")
# graph.add_edge("increment","route")
graph.add_conditional_edges("increment",
                            route,
                            {
                                "increment":"increment",
                                "stop":END
                            }
                           )

    
```

```python
app = graph.compile()
```

```python
print(app.get_graph().draw_ascii())
```
